```python
df['cuisine'].fillna("Unknown", inplace=True)
df['course'].fillna("Unknown", inplace=True)
df['diet'].fillna("Unknown", inplace=True)
```

```python
df['ingredients'] = df['ingredients'].str.replace('\n', '')
df['instructions'] = df['instructions'].str.replace('\n', '')
df['prep_time'] = df['prep_time'].str.replace('Total in ', '')
df['prep_time'] = df['prep_time'].str.replace(' M', '')
df['prep_time'] = df['prep_time'].astype(float)
```

```python
df['ingredients'] = df['ingredients'].str.replace('\n', '')
df['instructions'] = df['instructions'].str.replace('\n', '')

# User Input
available_ingredients = input("Enter the available ingredients (comma-separated): ").split(',')
time_allocation = float(input("Enter the time you can allocate (in minutes): "))
diet_preference = input("Enter your diet preference (veg or non-veg): ")

# Rule-Based Matching
matching_dishes = df[
    (df['ingredients'].str.contains('|'.join(available_ingredients), case=False)) &
    (df['prep_time'].apply(lambda x: float(x) <= time_allocation)) &
    (df['diet'] == diet_preference)
]

# Recommendation
if len(matching_dishes) > 0:
    recommended_dish = matching_dishes.sample(1)
    print("Recommended Dish:")
    print("Dish name:", recommended_dish['name'].values[0])
    print("Image URL:", recommended_dish['image_url'].values[0])
    print("Ingredients:", recommended_dish['ingredients'].values[0])
    print("Instructions:", recommended_dish['instructions'].values[0])
else:
    print("No matching dish found.")
```

```python
import tkinter as tk
from PIL import ImageTk, Image
import pandas as pd
import requests
from io import BytesIO
import random
from tkinter import *  

# Load and preprocess the dataset

df['ingredients'] = df['ingredients'].str.replace('\n', '')
df['instructions'] = df['instructions'].str.replace('\n', '')

# Function to recommend dishes
def recommend_dishes():
    # Clear previous recommendations and error messages
    recommendation_label.configure(text="")
    error_label.configure(text="")

    available_ingredients = ingredients_entry.get()
    time_allocation = time_entry.get()
    diet_preference = diet_variable.get()
    cuisine_preference = cuisine_variable.get()

    # Validate inputs
    if not available_ingredients or not time_allocation:
        error_label.configure(text="Please provide all the required inputs.")
        return

    try:
        time_allocation = float(time_allocation)
    except ValueError:
        error_label.configure(text="Invalid time allocation. Please enter a numeric value.")
        return

    # Rule-Based Matching
    matching_dishes = df[
        (df['ingredients'].str.contains('|'.join(available_ingredients.split(',')), case=False)) &
        (df['prep_time'].apply(lambda x: float(x) <= time_allocation)) &
        (df['diet'] == diet_preference) &
        (df['cuisine'] == cuisine_preference)
    ]

    # Recommendation
    if len(matching_dishes) > 0:
        try:
            recommended_dishes = matching_dishes.sample(min(len(matching_dishes), 3))  # Select up to 3 random dishes
            for i, dish in enumerate(recommended_dishes.iterrows()):
                index, dish_data = dish
                image_url = dish_data['image_url']
                ingredients = dish_data['ingredients']
                instructions = dish_data['instructions']

                # Update the image
                response = requests.get(image_url)
                response.raise_for_status()  # Check for HTTP errors
                img = Image.open(BytesIO(response.content))
                img = img.resize((200, 200))  # Resize the image for display
                photo = ImageTk.PhotoImage(img)
                image_labels[i].configure(image=photo)
                image_labels[i].image = photo

                # Update the recommendation details
                recommendation_labels[i].configure(
                    text=f"Dish {i+1}:\n\nIngredients: {ingredients}\n\nInstructions: {instructions}")
        except (requests.RequestException, IOError) as e:
            error_label.configure(text="Error occurred while fetching images.")
        except Exception as e:
            error_label.configure(text="An unexpected error occurred.")
    else:
        recommendation_label.configure(text="No matching dishes found.")

# Function to clear the input fields and error messages
def clear_inputs():
    ingredients_entry.delete(0, tk.END)
    time_entry.delete(0, tk.END)
    diet_variable.set("veg")
    cuisine_variable.set("Indian")
    recommendation_label.configure(text="")
    error_label.configure(text="")

# Create the main window
window = tk.Tk()
window.title("Dish Recommendation System")

# Create and place the input widgets
ingredients_label = tk.Label(window, text="Available Ingredients:")
ingredients_label.pack()
ingredients_entry = tk.Entry(window)
ingredients_entry.pack()

time_label = tk.Label(window, text="Time Allocation (minutes):")
time_label.pack()
time_entry = tk.Entry(window)
time_entry.pack()

diet_label = tk.Label(window, text="Diet Preference:")
diet_label.pack()
diet_variable = tk.StringVar(window, "Vegetarian")
diet_options = ["Vegetarian", "Non Vegeterian"]
diet_dropdown = tk.OptionMenu(window, diet_variable, *diet_options)
diet_dropdown.pack()

cuisine_label = tk.Label(window, text="Cuisine Preference:")
cuisine_label.pack()
cuisine_variable = tk.StringVar(window, "Indian")
cuisine_options = df['cuisine'].unique()
cuisine_dropdown = tk.OptionMenu(window, cuisine_variable, *cuisine_options)
cuisine_dropdown.pack()

submit_button = tk.Button(window, text="Submit", command=recommend_dishes)
submit_button.pack()

clear_button = tk.Button(window, text="Clear", command=clear_inputs)
clear_button.pack()

# Create and place the recommendation labels
recommendation_label = tk.Label(window, text="", wraplength=300)
recommendation_label.pack()

error_label = tk.Label(window, text="", wraplength=300, fg="red")
error_label.pack()

image_labels = []
recommendation_labels = []
for i in range(3):
    image_label = tk.Label(window)
    image_label.pack()
    image_labels.append(image_label)

    recommendation_label = tk.Label(window, text="", wraplength=300)
    recommendation_label.pack()
    recommendation_labels.append(recommendation_label)

window.mainloop()
```
